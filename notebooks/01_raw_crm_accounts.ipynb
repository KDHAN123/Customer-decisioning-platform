```python
assert (
    accounts["account_created_at"]
    >= pd.Timestamp("2024-06-01")
).all()

assert (
    accounts["account_created_at"]
    <= pd.Timestamp("2026-07-19 23:59:59")
).all()
```

```python
assert (
    accounts.loc[
        accounts["subscription_status"].isin(
            ["Never Subscribed", "Cancelled"]
        ),
        "current_plan",
    ]
    == "None"
).all()

print("All account validation checks passed.")
```
